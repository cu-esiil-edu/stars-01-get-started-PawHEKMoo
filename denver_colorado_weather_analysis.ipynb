```python
# Import pandas
import pandas as pd
```

```python
# Data for Yellowstone Lake

ncei_url = (
    'https://www.ncei.noaa.gov/access/services/data/v1'
    '?dataset=daily-summaries'
    '&dataTypes=TOBS,PRCP'
    '&stations=USW00094173'
    '&startDate=1949-10-01'
    '&endDate=2024-05-03'
    '&includeStationName=true'
    '&includeStationLocation=1'
    '&units=standard')
ncei_url
```

```python
# creating data frame for Yellowstone Lake
yellowstone_df = pd.read_csv(
  ncei_url,
  index_col='DATE',
  parse_dates=True,
  na_values=['NaN'])
yellowstone_df
```

```python
yellowstone_df = yellowstone_df[['PRCP','TOBS']]
yellowstone_df

# Resample the data to look at yearly mean values
ysl_annual_avg_precipitation = yellowstone_df.resample('YS').mean()
ysl_annual_avg_precipitation
```

```python
# Plot the data using .plot
yellowstone_df.plot(
    y='PRCP',
    title='Yellowstone Lake Precipitation',
    xlabel='Date',
    ylabel='Precipiation in Inches',
    legend=False,
    color='lightblue',
    figsize=(10,5))
yellowstone_df
```
